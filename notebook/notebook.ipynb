```python
dfs_2018 = fct_read_json_nested(root_json_2018)

df_stadiums = dfs_2018['stadiums']
df_tvchannels = dfs_2018['tvchannels']
df_teams = dfs_2018['teams']
df_matches = dfs_2018['matches']
```

```python
df_stadiums.head()
list_wanted_columns = [
    "id",
    "name",
    "city"
    ]
df_stadiums = df_stadiums[list_wanted_columns]
df_stadiums.head()
```

```python
df_tvchannels.head()
list_wanted_columns = [
    "id",
    "name",
    "country",
    "lang"
]
df_tvchannels = df_tvchannels[list_wanted_columns]
df_tvchannels = (df_tvchannels
                 .explode('lang')           # √©clate la liste
                 .rename(columns={'lang': 'lang'})
                 .reset_index(drop=True)
) 
df_tvchannels.head()
```

```python
df_teams.head()
list_wanted_columns = [
    "id",
    "name",
    "fifaCode",
    "emojiString"
]
df_teams = df_teams[list_wanted_columns]
df_teams.head()
```

```python
df_matches.head()
list_wanted_columns = [
    "name",
    "type",
    "home_team",
    "away_team",
    "home_result",
    "away_result",
    "date",
    "stadium",
    "channels",
    "group",
    "stage",  # a verifier avec type
    "winner"
]
df_matches = df_matches[list_wanted_columns]
df_matches.head()
  
```

```python
df_matches =  (
    df_matches
    .explode('channels')           # √©clate la liste
    .rename(columns={'channels': 'channelid'})
    .reset_index(drop=True)
)
df_matches.head()
```

```python
df_team_home = df_teams.add_prefix('home_')
df_team_away = df_teams.add_prefix('away_')
df_stadiums= df_stadiums.add_prefix('stadium_')
df_tvchannels= df_tvchannels.add_prefix('channel_')
df_team_home.head()
# df_team_away.head()
```

```python
df_2018_merged = (df_matches
                  .merge(df_stadiums, left_on='stadium', right_on='stadium_id', how='left')
                  .merge(df_tvchannels, left_on='channelid', right_on='channel_id', how='left')
                  .merge(df_team_home, left_on='home_team', right_on='home_id', how='left')
                  .merge(df_team_away, left_on='away_team', right_on='away_id', how='left')
                  )
df_2018_merged.head()
```

```python
list_final_columns = [
    'name',
    'type',
    'home_name',
    'away_name',
    'home_result',
    'away_result',
    'date',
    'stadium_name',
    'stadium_city',
    'type',
    'stage',
    'group',
    'winner',
    'channel_name',
    'channel_country',
    'channel_lang',
    'home_fifaCode',
    'away_fifaCode',
    'home_emojiString',
    'away_emojiString'
]
df_2018_final = df_2018_merged[list_final_columns]
df_2018_final.head()
    
```
